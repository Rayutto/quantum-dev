```python
from qutip import coherent, wigner
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha = 5.0
dim = 30
xvec = np.linspace(-10, 10, 500)

rho = coherent(dim, alpha)
W = wigner(rho, xvec, xvec)
```

```python
plt.contourf(xvec, xvec, W, 100, cmap='magma')
plt.title("Wigner function of |")
plt.xlabel("Re()")
plt.ylabel("Im()")
plt.colorbar()
plt.show()
```
